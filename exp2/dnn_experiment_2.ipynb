```python
import numpy as np
from numpy import random
```

```python
dataset = []
for i in range(100):
    dataset.append([])

for i in range(100):
    for j in range(100):
        dataset[i].append(random.randint(100))

for i in range(100):
    print(dataset[i])

y = random.randint(100)

print(y)
```

```python
trgt = []
w = []
w1 = []

# initialized weight values for the first data in dataset

print(f'Enter {len(dataset)} target variables (1-100) : ')
for i in range(len(dataset)):
    t = random.randint(0, 100)
    trgt.append(t)

alpha = float(input('Enter learning parameter : '))

b = float(input('Enter bias weight (0-10) : '))
n = b

print(f'Enter {len(dataset[0])} initial weights (0-10) : ')
for i in range(len(dataset[0])):
    w1data = random.randint(0, 10)
    w1.append(w1data)
```

```python
d = dataset[0]
for i in range(len(d)):
    n += d[i] * w1[i]

a = []
if n > y:
    a.append(1)
else:
    a.append(0)
```

```python
for i in range(len(d)):
    for j in range(len(a)):
        w.append(round(w1[j] + alpha * (trgt[j] - a[j]) * d[i],2))

print("Updated weights:", w)
```
